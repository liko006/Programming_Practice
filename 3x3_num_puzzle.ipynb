```python
import numpy as np
import copy

'''
목표 array 배열의 모양
[[0,1,2],
 [3,4,5],
 [6,7,8]]
'''
goal = [[0,1,2],
        [3,4,5],
        [6,7,8]]
goal = np.array(goal)

# state needs to be in 1x9 array shape
def getInvCount(state):

    inv_count = 0; 
    for i in range(9):
        for j in range(i+1,9):
            if state[i] > state[j] and state[i] != 0 and state[j] != 0: 
                 inv_count += 1 
    return inv_count 
  
# whether state is solvable or not
def is_solvable(state): 

    # Count inversions in given 8 puzzle 
    invCount = getInvCount(state); 
  
    # return true if inversion count is even. 
    return (invCount%2 == 0); 

# check state is goal
def is_goal(state):
    if np.array_equal(state, goal):
        return True
    else:
        return False

# check state is same
def is_same(state1, state2):
    if np.array_equal(state1, state2):
        return True
    else:
        return False
    
# move number left
def move_left(state, trace):
    new_state = copy.deepcopy(state)
    new_trace = copy.deepcopy(trace)
    for i in range(3):
        for j in [0,1]:
            if state[i][j] == 0:
                new_state[i][j] = state[i][j+1]
                new_state[i][j+1] = 0
                new_trace.append('Left')
                break
    return [new_state, new_trace]

# move number right
def move_right(state, trace):
    new_state = copy.deepcopy(state)
    new_trace = copy.deepcopy(trace)
    for i in range(3):
        for j in [1,2]:
            if state[i][j] == 0:
                new_state[i][j] = state[i][j-1]
                new_state[i][j-1] = 0
                new_trace.append('Right')               
                break
    return [new_state, new_trace]

# move number up
def move_up(state, trace):
    new_state = copy.deepcopy(state)
    new_trace = copy.deepcopy(trace)
    for i in [0,1]:
        for j in range(3):
            if state[i][j] == 0:
                new_state[i][j] = state[i+1][j]
                new_state[i+1][j] = 0
                new_trace.append('Up')
                break
    return [new_state, new_trace]

# move number down                
def move_down(state, trace):
    new_state = copy.deepcopy(state)
    new_trace = copy.deepcopy(trace)
    for i in [1,2]:
        for j in range(3):
            if state[i][j] == 0:
                new_state[i][j] = state[i-1][j]
                new_state[i-1][j] = 0
                new_trace.append('Down')
                break
    return [new_state, new_trace]

# check the avaliable moves in the given state and return all possible states
def next_states(prev_state):
    
    aval_row, aval_col = np.where(prev_state[0]==0)
    after_state = []
    
    if aval_row == 0 and aval_col == 0:
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 0 and aval_col == 1:
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state

    if aval_row == 0 and aval_col == 2:
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 1 and aval_col == 0:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 1 and aval_col == 1:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 1 and aval_col == 2:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_up(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 2 and aval_col == 0:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 2 and aval_col == 1:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_left(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
    if aval_row == 2 and aval_col == 2:
        after_state.append([move_down(prev_state[0], prev_state[1]), prev_state])
        after_state.append([move_right(prev_state[0], prev_state[1]), prev_state])
        return after_state
    
# give starting state   
temp = np.arange(0,9)
np.random.shuffle(temp)
initial = temp.reshape(3,3)
check_state = initial.reshape(9)
trace = []

start = [[initial, trace],]

state_list = []
state_list.append(start)

# run the programme
if is_solvable(check_state):

    i=0
    while True:

        current_state = state_list[i]

        if is_goal(current_state[0][0]):
            print('goal!')
            print(initial)
            print(current_state[0][1])
            print(goal)
            break
        
        next_state_list= []
        
        temp_list = next_states(current_state[0])
        
        for temp in temp_list:
            count = 0
            for j in range(len(state_list)):
                if is_same(temp[0][0], state_list[j][0][0]):
                    count += 1
            
            if count == 0:
                next_state_list.append(temp)
                        
        state_list = state_list + next_state_list
   
        i = i+1
        
        if i == len(state_list):
            break

else:
    print('Not Solvable')
```
